# Sankey Diagram Visualization for Flow Decomposition

This notebook demonstrates the new Sankey diagram visualization feature in flowpaths.

Sankey diagrams show flows through a network, where:
- Each node in the graph appears as a labeled node in the diagram
- Each path in the decomposition is shown as a colored flow
- The width of each connection represents the path weight
- The diagram is interactive (hover, zoom, pan)

```python
import networkx as nx
import flowpaths as fp
from flowpaths.utils import graphutils
```

## Example 1: Simple Flow Network

```python
# Create a simple DAG with flow values
G = nx.DiGraph()
G.add_edge('s', 'a', flow=10)
G.add_edge('s', 'b', flow=5)
G.add_edge('a', 'c', flow=6)
G.add_edge('a', 'd', flow=4)
G.add_edge('b', 'c', flow=3)
G.add_edge('b', 'd', flow=2)
G.add_edge('c', 't', flow=9)
G.add_edge('d', 't', flow=6)

print("Graph created with", G.number_of_nodes(), "nodes and", G.number_of_edges(), "edges")
```

```python
# Compute minimum flow decomposition
solver = fp.MinFlowDecomp(G, flow_attr='flow')
solver.solve()
solution = solver.get_solution()

paths = solution['paths']
weights = solution['weights']

print(f"Found {len(paths)} paths:")
for i, (path, weight) in enumerate(zip(paths, weights), 1):
    print(f"  Path {i}: {' → '.join(path)} (weight: {weight})")
```

### Visualize as Sankey Diagram

The diagram will appear interactively below. You can:
- Hover over nodes and links to see details
- Drag to rearrange (if enabled)
- Zoom and pan

```python
# This will display the Sankey diagram inline and save it as HTML
graphutils.draw(
    G,
    filename="example1_sankey",  # Will save as example1_sankey.html
    flow_attr='flow',
    paths=paths,
    weights=weights,
    draw_options={
        "style": "sankey"
    }
)
```

## Example 2: Larger Network

Let's try a more complex example with more paths.

```python
# Create a larger DAG
G2 = nx.DiGraph()
edges = [
    ('s', 'a', 15),
    ('s', 'b', 10),
    ('s', 'c', 8),
    ('a', 'd', 8),
    ('a', 'e', 7),
    ('b', 'e', 6),
    ('b', 'f', 4),
    ('c', 'f', 5),
    ('c', 'g', 3),
    ('d', 'h', 8),
    ('e', 'h', 10),
    ('e', 'i', 3),
    ('f', 'i', 5),
    ('f', 't', 4),
    ('g', 't', 3),
    ('h', 't', 18),
    ('i', 't', 8),
]
for u, v, w in edges:
    G2.add_edge(u, v, flow=w)

print("Graph created with", G2.number_of_nodes(), "nodes and", G2.number_of_edges(), "edges")
```

```python
# Compute decomposition
solver2 = fp.MinFlowDecomp(G2, flow_attr='flow')
solver2.solve()
solution2 = solver2.get_solution()

paths2 = solution2['paths']
weights2 = solution2['weights']

print(f"Found {len(paths2)} paths:")
for i, (path, weight) in enumerate(zip(paths2, weights2), 1):
    print(f"  Path {i}: {' → '.join(path)} (weight: {weight})")
```

```python
# Visualize the larger network
graphutils.draw(
    G2,
    filename="example2_sankey",
    flow_attr='flow',
    paths=paths2,
    weights=weights2,
    draw_options={
        "style": "sankey"
    }
)
```

## Notes

- The Sankey diagram requires the graph to be acyclic (DAG)
- Each path gets a different color (automatically assigned)
- Both HTML (interactive) and static image (PDF) files are generated automatically
